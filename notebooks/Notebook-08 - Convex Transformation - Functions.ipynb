# Non-linear Transformations of Functions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import kurtosis as scipy_kurtosis
```

### Transformation functions

##### Functions

```python
concave_convex_fn = lambda x: np.abs(x)*x 

def convex_concave_fn(x):
    
    is_positive = x>=0
    sign = is_positive+~is_positive*-1
    
    sqrt_of_abs = np.sqrt(np.abs(x))
    
    y = sqrt_of_abs*sign
    
    return y

def logistic_fn(x):
    
    numerator = 1
    denominator = (1+ (1/np.exp(x)) ) 
    y = numerator/denominator
    
    return y
```

##### Concave-Convex
i.e. _N_ shaped. Sort of like _logit and probit_ but not the same function.

```python
x = np.linspace(-10,10,10)
y = concave_convex_fn(x) 
assert (concave_convex_fn(x) == x*np.abs(x)).all() # Check we've defined the function as expected
```

```python
ax = plt.scatter(x,y, marker='*')
ax = plt.plot(x,y)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.axhline(color='k', linewidth=1)
plt.axvline(color='k', linewidth=1)
plt.show()
```

##### Convex-Concave I
i.e. _S_ shaped

```python
old_x, old_y = x.copy(), y.copy()
```

```python
x = np.linspace(-10,10,10)
y = convex_concave_fn(x) 
assert (convex_concave_fn(old_y) == old_x).all() # Check we've defined the function as expected
```

```python
ax = plt.scatter(x,y, marker='*')
ax = plt.plot(x,y)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.axhline(color='k', linewidth=1)
plt.axvline(color='k', linewidth=1)
plt.show()
```

##### Convex-Concave II
The Logistic Function

```python
x = np.linspace(-10,10,10)
y = logistic_fn(x)
```

```python
ax = plt.scatter(x,y, marker='*')
ax = plt.plot(x,y)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.axhline(color='k', linewidth=1)
plt.axvline(color='k', linewidth=1)
plt.show()
```
