# About
Some plots of an Exponential Distribution. 
* Plot probability density (PDF)
* Plot survival function (1 - CDF)

For the Survival Function I show log axes as well as linear ones.
* Standard axes (exponential is convex)
* Loglinear axes (exponential is linear)
* Loglog axes (exponential is concave)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

### Get Datapoints
Grab discrete datapoints for an exact exponential distribution.

```python
# Generate 100 Points
x = np.linspace(-2, 10, 100)       # 100 equally spaced x locations
pdf = expon.pdf(x, loc=0, scale=1) # 100 corresponding probability densities
survival = 1-expon.cdf(x, loc=0, scale=1) # 100 corresponding survival probabilities
```

# Plot PDF

```python
# Plot the probability density
fig, ax = plt.subplots()
ax.plot(x, pdf)
ax.set(xlabel='x', ylabel='PDF(x)',
       title='Exponential Distribution PDF')
plt.show()
```

Next I plot the survival function `P(X>x)`:

```python
# Plot Survival Function (Linear Axes)
fig, ax = plt.subplots()
ax.plot(x, survival)
ax.set(xlabel='x', ylabel='P(>x)',
       title='Survival Function (Linear Axes)')
plt.show()
```

```python
# Plot Survival Function (Semilog Axes)
fig, ax = plt.subplots()
ax.semilogy(x, survival)
ax.set(xlabel='x', ylabel='P(>x)',
       title='Survival Function (Semilog Axes)')
plt.show()
```

Note: the exponential distribution is linear in a semi-log plot whereas power laws are linear on a log-log plot.

To be thorough I'll regenerate datapoints for the loglog scale. There is no `x=0` on the loglog scale so I'll start with `x=0.1`

```python
# Loglog Plot
# Regenerate points to avoid `x=0`, this point doesn't exist on the log scale.
x = np.linspace(0.1, 10, 100)     
survival = 1-expon.cdf(x, loc=0, scale=1)
```

```python
# Plot Survival Function (Loglog Axes)
fig, ax = plt.subplots()
ax.loglog(x, survival)
ax.set(xlabel='x', ylabel='P(>x)',
       title='Survival Function (Loglog Axes)')
plt.show()
```
