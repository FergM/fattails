# About
This notebook shows how we can fatten the tails of the normal distribution. 

The tails are fattened by mixing between two distributions of different standard deviations (1+a and 1-a respectively). Each distribution gets selected 50% of the time. The result is an overall distribution which is fatter than a standard normal distribution with std=1.

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import multivariate_normal
```

###  Plot Simple Gaussian

```python
# Generate 100 Points
x = np.linspace(-4, 4, 100)                   # 100 equally spaced x locations
y = norm.pdf(x, loc=0, scale=1) # 100 corresponding probability densities
```

```python
# Plot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel='x', ylabel='pdf(x)',
       title='Normal Distribution PDF')
#ax.grid()
plt.show()
```

### Plot Fattened Gaussian

```python
a = 0.5 # Used to adjust the standard deviation
high_std = 1+a
low_std = 1-a

x = np.linspace(-5, 5, 100)
y = norm.pdf(x, loc=0, scale=1)
y_fat = 0.5*norm.pdf(x, loc=0, scale=high_std) + 0.5*norm.pdf(x, loc=0, scale=low_std)
```

```python
# Plot
fig, ax = plt.subplots()

ax.plot(x, y, label='Gaussian pdf')
ax.plot(x, y_fat, label='Gaussian pdf Mix')

ax.set(xlabel='x', ylabel='pdf(x)', title='Normal Distribution PDF')
ax.legend(loc='best', frameon=False)
#ax.grid()
plt.show()
```
