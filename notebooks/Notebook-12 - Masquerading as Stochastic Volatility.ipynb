```python
import numpy as np
import pandas as pd
```

```python
from scipy import stats
```

# Random Sample of Student t

```python
# Settings
dof = 3
location = 55
scale = 4
size=365*48
```

```python
# Statistics
mean, var, skew, kurt = stats.t.stats(dof, location, scale, moments='mvsk')
std = np.sqrt(var)
#Note: var = dof*scale**2
print('mean',mean)
print('std',int(std))
```

```python
# Sample
r = stats.t.rvs(dof, 55, 4, size=size)
```

```python
# Sample Plot
sr = pd.Series(r)
sr.plot(figsize=[7,5])
```

# Rolling Volatility

```python
sr.rolling(30).std().plot(figsize=[15,5])
ax = sr.rolling(30).std().plot(figsize=[15,5], title='Monthly Rolling Vol - One Year')
ax.axhline(sr.std(), color='k')
ax.axhline(std, color='g')
```

```python
subsample = sr.iloc[0:60*48]
ax = subsample.rolling(30).std().plot(figsize=[15,5], title='Monthly Rolling Vol - Two Months')
ax.axhline(subsample.std())
ax.axhline(sr.std(), color='k')
ax.axhline(std, color='g')
```
