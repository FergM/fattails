```python
import numpy as np
import pandas as pd
from scipy.stats import kurtosis as scipy_kurtosis 
```

# Create Normal Dataset

```python
normal_array = np.random.normal(size=8000)
```

```python
df = pd.DataFrame(normal_array, columns = ['price'])
```

```python
df.head()
```

# Functions

```python
def moment(arr, i: int):
    """Calculate the ith moment of array like input a"""
    
    moment = ((arr-arr.mean())**i).sum() / len(arr)
    
    return moment
```

```python
def my_kurtosis(a):

    kurtosis = moment(a, 4) / moment(a, 2)**2
    
    return kurtosis
```

```python
# Print the Moments
for i in range(0,7):
    print(f'Moment {i} is: {"{:.0e}".format(moment(df.price, i))}')
```

# Analyse

```python
df.price.plot(figsize=[15,5])
```

# Moments

```python
# Second Moment
moment(df.price, 2)
```

```python
# Variance is Second Moment
df.price.var()
```

### Kurtosis

```python
df.price.kurtosis()
```

```python
scipy_kurtosis(df.price, bias=False)
```

```python
scipy_kurtosis(df.price)
```

```python
# Pearson Kurtosis
scipy_kurtosis(df.price, fisher=False)
```

```python
# My Kurtosis Function
my_kurtosis(df.price)
```
